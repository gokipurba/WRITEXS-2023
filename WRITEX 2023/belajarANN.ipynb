```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi bobot dan bias
weights = np.ones((3, 3))
biases = np.zeros(3)

# Dataset input dan output
input_data = np.array([[1, 1,1], [1, 0,0], [0, 1,0], [0, 0,0]])
output_data = np.array([[0, 1, 2], [1, 0, 1], [1, 0, 1], [0, 0, 0]])

# Parameter pelatihan
learning_rate = 0.1
epochs = 1000

# Fungsi aktivasi linier
def linear_activation(x):
    return x

# Fungsi untuk menghitung error
def calculate_error(predictions, targets):
    return np.mean(np.square(predictions - targets))

# List untuk menyimpan error setiap epoch
errors = []

# Pelatihan model
for epoch in range(epochs):
    # Forward pass
    predictions = np.dot(input_data, weights) + biases
    
    # Menghitung error
    error = calculate_error(predictions, output_data)
    errors.append(error)
    
    # Menghitung gradien
    gradient = 2 * (predictions - output_data) / len(input_data)
    
    # Update bobot dan bias menggunakan penurunan gradien stokastik
    weights -= learning_rate * np.dot(input_data.T, gradient)
    biases -= learning_rate * np.sum(gradient, axis=0)
    
    # Cetak error setiap 100 epoch
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Error = {error}")

# Plot konvergensi error
plt.plot(range(epochs), errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error Convergence')
plt.show()

# input_data = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
predictions = np.dot(input_data, weights) + biases
print(predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset input dan output
input_data = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
output_data = np.array([[0], [1], [1], [0]])

# Inisialisasi bobot dan bias
weights = np.ones((2, 1))
bias = np.zeros((1, 1))
learning_rate = 0.1

# Fungsi aktivasi linier
def linear_activation(x):
    return x

# Fungsi untuk menghitung error
def calculate_error(predictions, targets):
    return np.mean(np.square(predictions - targets))

# Pelatihan model
loss_history = []

for epoch in range(1000):
    # Forward pass
    predictions = linear_activation(np.dot(input_data, weights) + bias)
    
    # Menghitung error
    error = calculate_error(predictions, output_data)
    loss_history.append(error)
    
    # Menghitung gradien
    gradient = 2 * (predictions - output_data) / len(input_data)
    
    # Update bobot dan bias menggunakan penurunan gradien stokastik
    weights -= learning_rate * np.dot(input_data.T, gradient)
    bias -= learning_rate * np.sum(gradient)
    
    # Cetak error setiap 100 epoch
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Error = {error}")

# Mengujikan model
test_input = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
test_output = np.array([[0], [1], [1], [0]])

test_predictions = linear_activation(np.dot(test_input, weights) + bias)
test_error = calculate_error(test_predictions, test_output)
print(weights)
print(f"Test Error: {test_error}")
print("Predictions:")
print(test_predictions)

# Plot konvergensi loss
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Convergence")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Dataset input dan output
input_data = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
output_data = np.array([[0, 1], [1, 1], [1, 1], [0, 0]])

# Inisialisasi bobot dan bias
input_dim = input_data.shape[1]
hidden_dim = 5  # Jumlah neuron pada hidden layer
output_dim = output_data.shape[1]

weights1 = np.random.randn(input_dim, hidden_dim)
bias1 = np.zeros((1, hidden_dim))
weights2 = np.random.randn(hidden_dim, output_dim)
bias2 = np.zeros((1, output_dim))

learning_rate = 0.1

# Fungsi aktivasi sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Fungsi aktivasi linier
def linear_activation(x):
    return x

# Fungsi untuk menghitung error
def calculate_error(predictions, targets):
    return np.mean(np.square(predictions - targets))

# Pelatihan model
loss_history = []

# Pelatihan model
for epoch in range(10000):
    # Forward pass
    hidden_layer_output = sigmoid(np.dot(input_data, weights1) + bias1)
    # hidden_layer_output = linear_activation(np.dot(input_data, weights1) + bias1)
    predictions = linear_activation(np.dot(hidden_layer_output, weights2) + bias2)
    
     # Menghitung error
    error = calculate_error(predictions, output_data)
    loss_history.append(error)
    
    # Menghitung gradien
    gradient2 = 2 * (predictions - output_data) / len(input_data)
    gradient1 = np.dot(gradient2, weights2.T) * (hidden_layer_output * (1 - hidden_layer_output))
    
    # Update bobot dan bias menggunakan penurunan gradien stokastik
    weights2 -= learning_rate * np.dot(hidden_layer_output.T, gradient2)
    bias2 -= learning_rate * np.sum(gradient2, axis=0)
    weights1 -= learning_rate * np.dot(input_data.T, gradient1)
    bias1 -= learning_rate * np.sum(gradient1, axis=0)

# Mengujikan model
test_input = np.array([[1, 1], [1, 0], [0.1, 0], [0, 0]])
test_output = np.array([[0, 1], [1, 1], [1, 1], [0, 0]])

hidden_layer_output = sigmoid(np.dot(test_input, weights1) + bias1)
test_predictions = linear_activation(np.dot(hidden_layer_output, weights2) + bias2)
test_error = calculate_error(test_predictions, test_output)

print(f"Test Error: {test_error}")
print("Predictions:")
print(test_predictions)

# Plot konvergensi loss
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Convergence")
plt.show()
```

```python
# !pip install tensorflow --upgrade --user
# !pip install tensorflow==2.1.0
from keras.models import Sequential
from keras.layers import Flatten, Dense
import numpy as np
import keras
from keras.layers import Activation
# from keras.utils.generic_utils import get_custom_objects

#dataset
X_train = np.array([[1, 1], [1, 0], [0, 1], [0, 0]]).T
X_train = np.array([[1, 1], [0, 0], [0, 1], [0, 0]]).T
y_train = np.array([[1, 1], [0, 1], [0, 1], [0, 0]]).T

X_test=X_train
y_test=y_train
epoch=20
epochs = range(epoch)

def custom_activation(x):
  return x

# get_custom_objects().update({'linear': Activation(custom_activation)})

#model menggunakan Multilayer Neural Network
model1 = Sequential()
model1.add(Flatten())
model1.add(Dense(16,activation='relu'))
model1.add(Dense(4,activation='softmax'))

#Training
optimizer = keras.optimizers.Adam(lr=0.1)
model1.compile(optimizer=optimizer,loss='categorical_crossentropy',metrics=['acc'])
history1 = model1.fit(X_train,y_train,epochs=epoch,batch_size=1,validation_data=(X_test,y_test))
model1.save('my_model.h5')
model1.summary()

#Testing
model1.evaluate(X_test,y_test)

#Plot Loss
import matplotlib.pyplot as plt
loss1 = history1.history['loss']
val_loss1 = history1.history['val_loss']
plt.plot(epochs,loss1,'r',label='training loss ANN')
plt.plot(epochs,val_loss1,'b',label='validasi loss ANN')
plt.legend()
```
